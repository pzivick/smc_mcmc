Now we define the possible parameter combinations that can be fit for in the model. In total there are 15 possible parameters that could be fixed (with prior ranges given for all of them), but in practice most likely only ~9 or so will be fit at any one time. Maybe even fewer if it seems that some parameters are very well constrained and can be left fixed in the future. The full list is:
ra0, dec0, pmra, pmdec, vsys, thet, incl, dist0, didt, dtdt, rad0, rotvel, sigma_pmra, sigma_pmdec, sigma_vsys, phi_tid, theta_tid, vel_tid

In general, ra0, dec0, vsys, and dist0 will be assumed values. didt and dtdt are tied to the precession/nutation values, which we don't have great constraints for in the SMC, so will likely be ignored. And unless we have RV data available, we'll also be excluding sigma_vsys as a parameter. So the standard fitting will be for 11 parameters.

The full version of this, using only Gaia data, would allow ra0, dec0, vsys, and dist0 to vary as the calculation of the center of mass motion (and the resulting viewing perspective effect) does depend on those four so we do have some leverage over them.

For prior ranges:
- We know that pmra and pmdec for the SMC should be small, but it would still be good to allow for some exploration of the space. So -5 to 5 mas/yr. Linear should be fine.
    
- For theta and the inclination angle, we want the full possible range, so 0 to 360 and -90 to 90, respectively. Linear should be fine.
    
- For the radius of peak rotation, we will need to explore allowing a very small radius out to a large radius, to be robust, probably 0.01 kpc to 10 kpc. So -2, 1 in log space.
    
- For the rotation velocity, we'll need it to cover positive and negative, so we probably have to just do linear space, for around -50 to 50 km/s.
    
- The dispersions should be on the order of ~10 km/s, but to properly explore the space, we should probably allow it down to like 0.1 km/s and up to 100 km/s to be thorough. So -1 to 2 in log space. And this will be the same for both of them. That said, we're doing the dispersion in PM space, so 100 km/s at 60 kpc would be 0.35 mas/yr. 1 km/s dispersion would be 0.0035 mas/yr. So in PM space, we'll do from -3 (0.001 mas/yr) to 0 (1 mas/yr) in log space.
    
- For the tidal expansion angles, phi_tid and theta_tid, we want the full range for angle in the plane of the sky. We should also go -90 to 90 for the theta angle because we'll want to explore multiple orders for the magnitude of the expansion, which will make it difficult to use to flip the direction of the vector.
    
- For that tidal expansion magnitude, we'll want something small like 0.1 km/s/kpc up to something large like 100 km/s/kpc (which we don't expect). So in log space, -1 to 2.
    
- In the case of the central RA and Dec value, it should be within the visible body of the SMC, so we can limit that relatively easily and just stick to linear space, say 0 to 30 for RA and -75 to -65 for Dec
    
- For distance, it should be between 55 and 65 kpc. No evidence exists to suggest something outside of that.
    
- Finally, for vsys, while there are some differing values, it should be likely centered around 145 km/s. So we'll expand that a bit to, say, 120 km/s to 170 km/s. 



```python
def get_priors(perm):
    
    if perm == 'standard':
        nparams = 11
        params = ['pmra0', 'pmdec0', 'thet', 'incl', 'rad0', 'rotvel', 'sigma_pmra', 'sigma_pmdec', \
                  'phi_tid', 'theta_tid', 'vel_tid']
        priors = np.asarray([[-5., 5.], [-5., 5.], [0., 360.], [-90., 90.], [-2., 1.], [-50., 50.], 
                            [-2., 0.], [-2., 0.], [0., 360.], [-90., 90.], [-1., 2.]])
        
    elif perm == 'full':
        nparams = 15
        params = ['pmra', 'pmdec', 'thet', 'incl', 'rad0', 'rotvel', 'sigma_pmra', 'sigma_pmdec', \
                  'phi_tid', 'theta_tid', 'vel_tid', 'ra0', 'dec0', 'dist0', 'vsys']
        priors = np.asarray([[-5., 5.], [-5., 5.], [0., 360.], [-90., 90.], [-2., 1.], [-50., 50.], \
                            [-3., 0.], [-3., 0.], [0., 360.], [-90., 90.], [-1., 2.], \
                            [0., 30.], [-75., -65.],[55., 65.], 120., 170.])
        
    else:
        nparams = 11
        params = ['pmra', 'pmdec', 'thet', 'incl', 'rad0', 'rotvel', 'sigma_pmra', 'sigma_pmdec', \
                  'phi_tid', 'theta_tid', 'vel_tid']
        priors = np.asarray([[-5., 5.], [-5., 5.], [0., 360.], [-90., 90.], [-2., 1.], [-50., 50.], 
                            [-3., 0.], [-3., 0.], [0., 360.], [-90., 90.], [-1., 2.]])
        
    return priors, nparams, params
```

Now we have a small function that generates the initial states for the MCMC using the prior range in order to make sure that the full space is being properly explored from the jump.

```python
def create_initial(p0, priors):
    
    for a in range(len(p0[0])):
        p0[:,a] = (p0[:,a] * (priors[a][1] - priors[a][0])) + priors[a][[0]]
        
    return p0
```

```python
def calc_cov_wcs(data):

    cov = np.zeros((len(data), 2, 2))
    cov[:,0,0] = data['pmra_error']**2 
    cov[:,0,1] = data['pmra_error'] * data['pmdec_error'] * data['pmra_pmdec_corr']
    cov[:,1,0] = data['pmra_error'] * data['pmdec_error'] * data['pmra_pmdec_corr']
    cov[:,1,1] = data['pmdec_error']**2
        
    return cov
```

Now before we dive into discussing the proper motion model used, we will first discuss the viewing perspective correction that will be added into the proper motion model: (for now, see the attached pdf "smc_model_guide")

For the proper motion distribution, we assume a Gaussian likelihood for each population, which can be written as:

$\mathcal{L}_{\mathrm{pm}} = (2\pi)^{-1/2} (\det\Sigma)^{-1/2} \exp{\displaystyle[ -\frac{1}{2} (\mu - \overline{\mu})^{T} \Sigma^{-1} (\mu - \overline{\mu})}]$,

where $\mu$ is the proper motion of a given star, $\overline{\mu}$ is the systemic proper motion of the population, and $\Sigma$ is the covariance matrix for the sample, which can be written as:

$$\begin{bmatrix} E_1 + S_1 & 0_{2\times2} & ... & 0_{2\times2} \\ 0_{2\times2} & E_2 + S_2 & ... & 0_{2\times2} \\ ... & ... & ... & ... \\ 0_{2\times2} & 0_{2\times2} & ... & E_N + S_N \end{bmatrix}$$, 

where $E$ and $S$ are:

$ E_i = \begin{bmatrix}\epsilon^2_{i,x} & \rho_i \epsilon_{i,x} \epsilon_{i,y} \\ \rho_i \epsilon_{i,x} \epsilon_{i,y} & \epsilon^2_{i,y} \end{bmatrix}$ and $ S_i = \begin{bmatrix} \sigma^2_i(x_i) & 0 \\ 0 & \sigma^2_i(x_i)\end{bmatrix}$,

where $\epsilon$ is the proper motion uncertainty, $\rho$ is the correlation between proper motion in $x$ and $y$, and $\sigma$ is the internal dispersion of the given system.

For the code, given that these are small matrices, with the largest only being 2x2, we choose to manually calculate both the determinant and the matrix multiplication in the exponential to save on computational time. For the determinant, we can remember that for a 2x2 matrix, the determinant is:

$\Sigma = \begin{bmatrix}a & b \\ c & d\end{bmatrix}, \det(\Sigma) = a*d - b*c$

And for the multiplication in the exponential, we have two slightly different pieces that we'll need to calculate in the code first. First, we have the 2x1 vector that is the difference between the model $\mu$ and the $\mu$ of a given star, which we'll call $\mu_{\mathrm{diff}}$ with the components labeled as $x$ and $y$:

Defining $\mu_{\mathrm{diff}} = \begin{bmatrix}x \\ y\end{bmatrix}$

Our second piece that we need to calculate manually is the inverse matrix of $\Sigma$, which for a 2x2 matrix is:

$\Sigma^{-1} = \frac{\displaystyle 1}{\displaystyle \det(\Sigma)}\begin{bmatrix}d & -b \\ -c & a\end{bmatrix}$

For the code, given that we have a 3D array of the covariance matrices, we'll just apply the $1$ over the determinant to each of the components, so we'll refer to the modified components as $a^{\ast} = a\ / \det(\Sigma)$, etc.

So putting it all together, we have:

$\mu^T_{\mathrm{diff}}\ \Sigma^{-1}\ \mu_{\mathrm{diff}} = \begin{bmatrix}x & y\end{bmatrix}\ \begin{bmatrix}d^{\ast} & -b^{\ast} \\ -c^{\ast} & a^{\ast}\end{bmatrix}\ \begin{bmatrix}x \\ y\end{bmatrix}$

If we multiply all of these components out, we're left with this final expression:

$\mu^T_{\mathrm{diff}}\ \Sigma^{-1}\ \mu_{\mathrm{diff}} = x\ (d^{\ast} x - b^{\ast} y) + y\ (-c^{\ast} x + a^{\ast} y)$


```python
def pm_gauss_model_wcs(pmra, pmdec, cov, pmra0, pmdec0):

    diff = np.zeros((len(pmra), 2, 1))
    
    diff[:,0,0] = pmra - pmra0
    diff[:,1,0] = pmdec - pmdec0
    
    model = np.zeros((len(pmra)))
    
    covdet = cov[:,0,0]*cov[:,1,1] - cov[:,0,1]*cov[:,1,0]
    
    invcov = np.zeros((len(cov), 2, 2))
    
    invcov[:,0,0] = cov[:,0,0] / covdet
    invcov[:,0,1] = cov[:,0,1] / covdet
    invcov[:,1,0] = cov[:,1,0] / covdet
    invcov[:,1,1] = cov[:,1,1] / covdet
    
    model = covdet**(-0.5) * np.exp(-0.5*(diff[:,0,0]*(invcov[:,1,1]*diff[:,0,0]-invcov[:,0,1]*diff[:,1,0]) + \
                                          diff[:,1,0]*(-1.0*invcov[:,1,0]*diff[:,0,0]+invcov[:,0,0]*diff[:,1,0])))
    
    return model
```

With the different models defined, we now move to a function that will define the different model cases that we might want to test in the MCMC.

```python
def total_ln_like(theta, data, cov, perm, dist0, ra0, dec0, vsys):
    
    #data, cov, cov2, mu_x1, mu_y1, mu_x2, mu_y2, sigma_x2, sigma_y2 - PM
    # data, Rmax, ah, pa, ellip- spatial
    
    cov_i = np.copy(cov) #this line is so that the original covariance matrix doesn't get overwritten
    
    
## permutation for only the proper motion for likelihood
    if (perm == 'standard'):
        pmra0, pmdec0, thet, incl, rad0, rotvel, sigma_pmra, sigma_pmdec, \
                  phi_tid, theta_tid, vel_tid = theta
        
        cov_i[:,0,0] += (10**sigma_pmra)**2
        cov_i[:,1,1] += (10**sigma_pmdec)**2
        
        pmra, pmdec, rv = pmmodel(pmra0, pmdec0, dist0, vsys, ra0, dec0, np.deg2rad(data['RA']), np.deg2rad(data['DEC']), \
                       incl, thet, (10**rad0), rotvel, phi_tid, theta_tid, (10**vel_tid))
        
            
        # Calculate the PM likelihoods for each population
        model_like = pm_gauss_model_wcs(data['PMRA'], data['PMDEC'], cov, pmra, pmdec)
        
        like = np.sum(np.log(model_like))
        
        
## permutation for only the spherical spatial case
    else:
        pmra0, pmdec0, thet, incl, rad0, rotvel, sigma_pmra, sigma_pmdec, \
                  phi_tid, theta_tid, vel_tid = theta
        
        cov_i[:,0,0] += (10**sigma_pmra)**2
        cov_i[:,1,1] += (10**sigma_pmdec)**2
        
        pmra, pmdec, rv = pmmodel(pmra0, pmdec0, dist0, vsys, ra0, dec0, np.deg2rad(data['RA']), np.deg2rad(data['DEC']), \
                       incl, thet, (10**rad0), rotvel, phi_tid, theta_tid, (10**vel_tid))
        
            
        # Calculate the PM likelihoods for each population
        model_like = pm_gauss_model_wcs(data['PMRA'], data['PMDEC'], cov, pmra, pmdec)
        
        like = np.sum(np.log(model_like))
        
        
    return like
```

With the various options for how to construct the likelihood finished, we turn to setting up the functions to handle the priors and check whether values being checked by the walkers are allowed by the priors: 

```python
# function to check that all priors are within flat prior range. For now it's set up assuming flat priors. Will
# examine how to modify the setup to include the ability to assign Gaussian priors.

def ln_prior(theta, perm):
    if (perm == 'standard'):
        pmra0, pmdec0, thet, incl, rad0, rotvel, sigma_pmra, sigma_pmdec, phi_tid, theta_tid, vel_tid = theta
        
        lp = 0. if (-5. < pmra0 < 5.) and (-5. <= pmdec0 <= 5.) and (0. <= thet <= 360.) \
        and (-90. < incl < 90.) and (-2. < rad0 < 1.) and (-50. < rotvel < 50.) and (-2. < sigma_pmra < 0.) \
        and (-2. < sigma_pmdec < 0.) and (0. < phi_tid < 360.) and (-90. < theta_tid < 90.) \
        and (-1. < vel_tid < 2.) else -np.inf
        
        
    else:
        lp = -np.inf

        
    return lp
        
        
#def total_ln_like(theta, data, perm='PM_ellip_spat', cov=None, cov2=None, Rmax=None): guesses
def ln_prob(theta, data, cov, perm, dist0, ra0, dec0, vsys):
        
    lp = ln_prior(theta, perm)
    if not np.isfinite(lp):
        return -np.inf
    try:
        lnprobval = lp + total_ln_like(theta, data, cov, perm, dist0, ra0, dec0, vsys)
    except ValueError: # NaN value case
        lnprobval = -np.inf
        
    return lnprobval
```

Now we set up the function that will actually run the MCMC:

```python
def run_MCMC(data, nparams, perm='standard', ra0=np.deg2rad(13.0), dec0=np.deg2rad(-71.0), dist0=60.0, \
            vsys=145., priors=None, nw=40, ns=1000):
    
    # Calculate the covariance matrices
    cov = calc_cov_wcs(data)

    # Set the number of dimensions and number of walkers desired
    ndim, nwalkers = nparams, nw
    
    # Create the initial state used to sample the space
    p0 = np.random.rand(nwalkers, ndim)
    
    p0 = create_initial(p0, priors)
    
    # Create the sampler itself
    sampler = emcee.EnsembleSampler(nwalkers, ndim, ln_prob, args=[data, cov, perm, dist0, ra0, dec0, vsys])
    
    print("Running burn-in...")
    state = sampler.run_mcmc(p0, 100, progress=True)
    sampler.reset()
    sampler.run_mcmc(state, ns, progress=True);
            
    return sampler, cov
```
